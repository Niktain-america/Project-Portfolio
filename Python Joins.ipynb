```python
import pandas as pd
```

```python
emp=pd.read_csv('BioData.csv')
dep=pd.read_csv('Department_ID.csv')
```

```python
emp
```

```python
dep
```

```python
innerjoin=emp.merge(dep,left_on='Department_ID',right_on='ID',how='inner')
innerjoin
```

```python
leftjoin=emp.merge(dep,left_on='Department_ID',right_on='ID',how='left')
leftjoin
```

```python
rightjoin=emp.merge(dep,left_on='Department_ID',right_on='ID',how='right')
rightjoin
```

```python
fulljoin=emp.merge(dep,left_on='Department_ID',right_on='ID',how='outer')
fulljoin
```
