```python
import pandas as pd

df =pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data.csv")
df.head()
```

```python
df=df.drop(columns='Total')
```

```python
df.info()
```

```python
df['Total']=df['Units']*df['Unit Cost']
```

```python
df.head()
```

```python
df.groupby('Rep').sum()
```

```python
df.groupby('Rep', as_index=False).sum()
```

```python
df.groupby('Rep', as_index=False)['Total'].sum()
```

```python
df.groupby('Rep', as_index=False)['Unit Cost'].sum()
```

```python
df.groupby('Rep', as_index=False)['Total'].agg('mean')
```

```python
df.groupby('Rep')['Total'].agg(['sum','mean','max'])
```

```python
df.groupby('Rep')[['Total','Unit Cost']].agg({'Total':['max','min'], 'Unit Cost':['mean','sum']})
```

```python
df.groupby(['Rep','Region']).mean()
```

```python
df.groupby(['Rep','Item']).sum()
```

```python
df.groupby(['Rep','Item']).mean()
```

```python
df.groupby(['Rep','Item'])['Units'].agg(['max'])
```

```python
df.groupby(['Rep','Item'])['Unit Cost'].agg(['max'])
```

```python
df.groupby(['Rep','Item'])['Units'].agg(['max','min'])
```

```python
df.groupby(['Rep','Item'])['Unit Cost'].agg(['max','min'])
```
