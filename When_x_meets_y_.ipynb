Here we import numpy to represent data as lists easily and quickly.
import tenserflow, from which we use keras.
keras is used for **implementing neural network**.
neural network is **just like brain** trying to find out
**whether is it female or male?** or whether is it good movie or bad movie?

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## **Define and compile the neural network. **

```python
model = tf.keras.Sequential([keras.layers.Dense(units=6, input_shape=[1])])
```


for compiling the neural network we use two function loss and optimizer.
loss **guesses answers against correct answers**. It measures **how well or badly it did**.
Optimizer is used to **make another guess**.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

# **Provide the data**
NumPy is a python library which **provides **lots of **array type data structures**. np.array[] is used to  **specify the value as an array in NumPy**.

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

# **To Train the neural network** .
This where it **learns the relationships between X's and Y's**.

```python
model.fit(xs, ys, epochs=500)
```

# **Use the Model**.
using **model.predict**

```python
print(model.predict([10.0]))
```
