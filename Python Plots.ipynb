```python
import matplotlib.pyplot as plt
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.scatter(x,y)
plt.show()
```

```python
x=[5,2,9,4,7]
y=[1,2,3,4,5]
plt.bar(x,y)
plt.show()
```

```python
y=[1,2,3,4,5,1,2,1,6,7,9,10]
plt.hist(y)
plt.show()
```
