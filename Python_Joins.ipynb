```python
emp=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ BioData.csv')
dep=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Department_ID.csv')
```

```python
emp
```

```python
dep
```

```python
innerjoin=dep.merge(emp, left_on='ID', right_on='Department_ID',how='inner')
innerjoin
```

```python
leftjoin=emp.merge(dep,left_on='Department_ID',right_on='ID',how='left')
leftjoin
```

```python
rightjoin=emp.merge(dep, left_on='Department_ID', right_on='ID', how='right')
rightjoin
```

```python
fulljoin=emp.merge(dep,left_on='Department_ID',right_on='ID',how='outer')
fulljoin
```
